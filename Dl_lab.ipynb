```python
# Scale the data
from sklearn.preprocessing import MinMaxScaler
df = yf.download('AMZN', start='2020-01-01', end='2024-01-01')

scaler = MinMaxScaler(feature_range=(0,1))
scaled_data = scaler.fit_transform(dataset)

scaled_data
```

```python
# Create the training data set
# Create the scaled training data set
train_data = scaled_data[0:int(training_data_len), :]
# Split the data into x_train and y_train data sets
x_train = []
y_train = []

for i in range(60, len(train_data)):
    x_train.append(train_data[i-60:i, 0])
    y_train.append(train_data[i, 0])
    if i<= 61:
        print(x_train)
        print(y_train)
        print()

# Convert the x_train and y_train to numpy arrays
x_train, y_train = np.array(x_train), np.array(y_train)

# Reshape the data
x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))
# x_train.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM

# Build the LSTM model
model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape= (60, 1)))
model.add(LSTM(50, return_sequences=False))

model.add(Dense(1))

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(x_train, y_train, batch_size=1, epochs=1)
```

```python
# Check the size of your original dataset
print("Original Data shape:", df.shape)

# Get closing price column and scale
dataset = df[['Close']]
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(dataset)

# Set training data length (e.g., 80% of data)
training_data_len = int(len(scaled_data) * 0.8)
print("Training data length:", training_data_len)

# Prepare test data
test_data = scaled_data[training_data_len - 60:, :]
print("Test data shape (with 60 padding):", test_data.shape)




# Create x_test and y_test
x_test = []
y_test = []

for i in range(60, len(test_data)):
    x_test.append(test_data[i - 60:i, 0])
    y_test.append(test_data[i, 0])

x_test = np.array(x_test)
y_test = np.array(y_test)
print("x_test shape before reshape:", x_test.shape)
print("y_test shape:", y_test.shape)

# Reshape for LSTM input
x_test = x_test.reshape((x_test.shape[0], x_test.shape[1], 1))
print("x_test shape after reshape:", x_test.shape)

# Make predictions
predictions = model.predict(x_test)
print("Predictions shape before reshape:", predictions.shape)

# Reshape predictions and y_test for inverse scaling
predictions = predictions.reshape(-1, 1)
y_test = y_test.reshape(-1, 1)

# Inverse transform
predictions = scaler.inverse_transform(predictions)
y_test_original = scaler.inverse_transform(y_test)


from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
mse = mean_squared_error(y_test_original, predictions)
print("MSE:", mse)


r2 = r2_score(y_test_original, predictions)
print("R-squared:", r2)

# Final RMSE
rmse = np.sqrt(np.mean((predictions - y_test_original) ** 2))
print("RMSE:", rmse)

mae = mean_absolute_error(y_test_original, predictions)
print("MAE:", mae)
```

```python
train = df[:training_data_len].copy()
valid = df[training_data_len:].copy()

# Assign predictions to the valid DataFrame
valid['Predictions'] = predictions.ravel()

# Plot the actual vs predicted
plt.figure(figsize=(14, 6))
plt.title('Model Predictions vs Actual Closing Prices')
plt.plot(train['Close'], label='Train')
plt.plot(valid['Close'], label='Actual')
plt.plot(valid['Predictions'], label='Predicted')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.show()
```

```python
from keras.callbacks import EarlyStopping

from sklearn.model_selection import train_test_split

# Split training data into actual training and validation sets
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.2, random_state=42)

early_stop = EarlyStopping(monitor='loss', patience=5)
history=model.fit(x_train, y_train, epochs=20, batch_size=32, callbacks=[early_stop])
plt.figure(figsize=(8,5))
plt.plot(history.history['loss'], marker='o', linestyle='-', color='teal')
plt.title('Model Loss During Training', fontsize=14)
plt.xlabel('Epoch', fontsize=12)
plt.ylabel('Loss (MSE)', fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()


history = model.fit(
    x_train, y_train,
    validation_data=(x_val, y_val),  # ðŸ‘ˆ This is necessary
    epochs=8,
    batch_size=32,
    callbacks=[early_stop]
)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training vs Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()
```
